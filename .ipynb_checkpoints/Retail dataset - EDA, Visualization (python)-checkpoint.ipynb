# Exploratory data analysis Retail dataset using python

## 1. Import modules/libraries

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load data

```python
# Read data
retail_ds = pd.read_excel("F:\Projects\Retail dataset\Online Retail.xlsx")
```

```python
# Null value analysis
retail_ds.info()
```

## 3. Data Preparation

```python
# Added a new column TotalSales to plot Sales summary
retail_ds['TotalSales'] = retail_ds['Quantity']*retail_ds['UnitPrice']
```

```python
retail_ds.head()
```

```python
# Total uniques countries in dataset
retail_ds['Country'].nunique()
```

```python
# grouping by invoice number to calculate total invoices raised from each country
invoice_count = retail_ds.groupby(["InvoiceNo","Country"])["Country"].count()

# converting the grouped output to a dataframe in order a plot a chart with more control
total_invoice_df = invoice_count.groupby(["Country"]).size().to_frame(name = 'TotalInvoiceCount').reset_index()
total_invoice_df
```

##### Country wise customer base, UK shares major customer base

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x="Country", y="TotalInvoiceCount", data=total_invoice_df)
# Align x tick labels
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment='right')
```

##### Country wise customer base excluding UK

```python
plt.figure(figsize=(16, 6))
ax = sns.barplot(x="Country", y="TotalInvoiceCount", data=total_invoice_df[total_invoice_df["Country"] != 'United Kingdom'])
# Align x tick labels
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment='right')
```
